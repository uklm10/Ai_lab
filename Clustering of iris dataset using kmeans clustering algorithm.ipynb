# Clustering of Iris Dataset using KMeans Clustering Algorithm

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
import pandas as pd
```

```python
data=pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=3, init="random", algorithm="full")
model.fit(data)
```

```python
from matplotlib import pyplot as plt
```

```python
print(model.labels_)
```

```python
plt.scatter(data.iloc[:, 1], data.iloc[:, 3], c=model.labels_, cmap='brg')
plt.xlabel(iris.feature_names[1])
plt.ylabel(iris.feature_names[3])
plt.show()
```

```python
from sklearn.metrics.cluster import silhouette_score
print(silhouette_score(data, model.labels_))
```

```python
k_range=range(2, 50)
score=[]
for k in k_range:
    model=KMeans(n_clusters=k, init="random", algorithm="full")
    model.fit(data)
    score.append(silhouette_score(data, model.labels_))
```

```python
plt.figure(figsize=(20, 10))
plt.bar(k_range, score)
plt.xticks(k_range)
plt.xlabel('Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```
